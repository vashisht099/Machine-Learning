# Univariate Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv', names = ['ID','Sepal Length in cm','Sepal Width in cm', 'Petal Length in cm', 'Petal Width in cm', 'Species'], header = 0)
```

```python
df.tail()
```

```python
df.shape
```

```python
setosa = df.loc[df['Species']=='Iris-setosa']
virginica = df.loc[df['Species']=='Iris-virginica']
versicolor = df.loc[df['Species']=='Iris-versicolor']
```

```python
plt.plot(setosa['Sepal Width in cm'],np.zeros_like(setosa['Sepal Width in cm']))
plt.plot(virginica['Sepal Width in cm'],np.zeros_like(virginica['Sepal Width in cm']))
plt.plot(versicolor['Sepal Width in cm'],np.zeros_like(versicolor['Sepal Width in cm']))
```

```python
# We conclude that flowers having sepal width less than 3.5 cm belongs to setosa species and above 3.5 and below 4 are virginicac and above 4 belongs to versicolor
```

# Bivariate Analysis

```python
sns.FacetGrid(df,hue='Species',height= 6).map(plt.scatter,'Petal Width in cm', 'Petal Length in cm').add_legend()
```

```python
sns.FacetGrid(df,hue="Species",height=5).map(plt.scatter,'Sepal Width in cm', 'Sepal Length in cm').add_legend()
```
