```python
# import the required libraries
import pandas as pd
import requests
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'}
response = requests.get('https://www.edureka.co/all-courses?page=1')
response
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, 'lxml')
print(soup.prettify())
```

```python
course_info = soup.find_all('div', class_ = 'course_card_info')
```

```python
course_name = []
learning_hours = []
batch_type = []
start_date = []
certificate = []
for i in course_info:
    course_name.append(i.find_all('h2', class_ = 'title')[0].text.strip())
course_name
```

```python
for i in course_info:
    get_lst = i.find_all('ul', class_ = 'you_will_get')
    for j in get_lst:
        certificate.append(j.find_all('li')[0].text.strip())
        learning_hours.append(j.find_all('li')[1].text.strip())
        batch_type.append(j.find_all('li')[2].text.strip())
```

```python
for i in course_info:
    try:
        start_date.append(i.find_all('div', class_ = 'date_starting')[0].text.strip())
    except:
        start_date.append('')
start_date
```

```python
df = pd.DataFrame({'course': course_name,'start_date': start_date, 'batch_type': batch_type, 'learning_hours': learning_hours, 'certificate': certificate})
df
```

```python
import numpy as np
#extracting information from all web pages 
data = pd.DataFrame()
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'}

#we have total of 10 pages
for i in range(1,11):
    response = requests.get('https://www.edureka.co/all-courses?page={}'.format(i))
    soup = BeautifulSoup(response.text, 'lxml')
    course_info = soup.find_all('div', class_ = 'course_card_info')
    course_name = []
    learning_hours = []
    batch_type = []
    start_date = []
    certificate = []
    for i in course_info:
        try:
            course_name.append(i.find_all('h2', class_ = 'title')[0].text.strip())
        except:
            course_name.append(np.nan)
    for i in course_info:
        get_lst = i.find_all('ul', class_ = 'you_will_get')
        for j in get_lst:
            try:
                certificate.append(j.find_all('li')[0].text.strip())
            except:
                certificate.append(np.nan)
            try:
                learning_hours.append(j.find_all('li')[1].text.strip())
            except:
                learning_hours.append(np.nan)
            try:
                batch_type.append(j.find_all('li')[2].text.strip())
            except:
                batch_type.append(np.nan)
    for i in course_info:
        try:
            start_date.append(i.find_all('div', class_ = 'date_starting')[0].text.strip())
        except:
            start_date.append(np.nan)
    df = pd.DataFrame({'course': course_name,'start_date': start_date, 'batch_type': batch_type, 'learning_hours': learning_hours, 'certificate': certificate})
    data = data.append(df,ignore_index = True)
```
